```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df_oscar = pd.read_csv('data/raw/full_data.csv', sep='\t')
df_oscar
```

```python
df_oscar['Class'].value_counts()
```

```python
# Removendo SciTech e Special
df_oscar = df_oscar[~df_oscar['Class'].isin(['SciTech', 'Special'])]
df_oscar
```

```python
# Oscar nao associando a filme
df_oscar[df_oscar['Film'].isna()].shape[0]
```

```python
# Removendo oscars nao associados a filme
df_oscar = df_oscar[~df_oscar['FilmId'].isna()]
```

```python
# Check NomId for 2024
((df_oscar['Year']!='2024') & (df_oscar['NomId'].str.contains('fake'))).sum()
```

```python
# Buscando filmes com mais indicações ao oscar
nominations_count = df_oscar['FilmId'].value_counts()
nominations_count
```

```python
df_oscar.loc[:, 'Winner'] = df_oscar['Winner'].fillna(0).astype(int)
df_oscar
```

```python
wins_count = df_oscar[['FilmId', 'Winner']].groupby('FilmId').sum().sort_values(by='Winner', ascending=False)
wins_count
```

```python
# Carregando os dados de filmes 
df_title = pd.read_csv('data/processed/title.tsv.gz', sep='\t', compression='gzip')
df_title
```

```python
df_oscar.loc[ ~df_oscar['FilmId'].isin(df_title['tconst']), ['Year', 'Film', 'CanonicalCategory', 'FilmId'] ]
```

```python
df_oscar.loc[ ~df_oscar['FilmId'].isin(df_title['tconst']), 'CanonicalCategory'].value_counts()
```

```python
df_title = df_title\
.merge(nominations_count, left_on='tconst', right_on='FilmId', how='inner')\
.merge(wins_count, left_on='tconst', right_on='FilmId', how='inner')\
.rename(columns={'count': 'oscarNominations', 'Winner': 'oscarWins'})

df_title
```

```python
df_title.to_csv('data/processed/title_oscar.tsv.gz', sep='\t', compression='gzip', index=False)
```
